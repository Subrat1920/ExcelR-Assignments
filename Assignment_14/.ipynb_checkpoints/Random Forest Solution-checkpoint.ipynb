```python
## function for checking the columns with outliers
def check_outlier_columns(data, threshold=1.5):
    out_cols = []
    for col in df.columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - threshold*(iqr)
        upper_bound = q3 + threshold*(iqr)
        if (df[col].max() > upper_bound) or (df[col].min() < lower_bound):
            out_cols.append(col)
    return out_cols

## let's check the function 
check_outlier_columns(df, threshold=1.5)
```

```python
## defining a variabe to take all the columns with maximum number of outliers
outlier_cols = ['RI', 'Al', 'Si', 'Ca', 'Ba', 'Fe']
```

```python
## function for removing the outliers
def remove_outliers(df, columns, threshold=1.5):
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - threshold * iqr
        upper_bound = q3 + threshold * iqr
        ## Filter rows within the bounds using element-wise logical AND
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
## applying the function to remove outliers
df = remove_outliers(df, outlier_cols, threshold=1.5)
```

### <span style='color:#6A9C89'>3. Data Pre-Processing</span>

```python
## splitting the data into dependent and independent variable
x = df.drop(columns='Type', axis=1)
y = df.Type
```

```python
## splitting the data into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
## applying feature scalling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### <span style='color:#6A9C89'>4. Random Forest Model Implementation</span>

```python
## importing random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

## random forest classifier parameters
rfc_params = {
    'n_estimators' : [x*100 for x in range(0, 10) if x%2==0],
    'criterion' : ["gini", "entropy", "log_loss"],
    'max_depth' : [x for x in range(0,10)],
    'min_samples_split' : [x for x in range(2, 10)] ,
    'min_samples_leaf' : [x for x in range(1, 10)],
    'max_features' : ["sqrt", "log2", None],
    'max_leaf_nodes' : [x for x in range(1, 10) if x%2!=0],
    'class_weight' : ["balanced", "balanced_subsample"],
    'ccp_alpha' : [0.1, 0.3, 0.5, 0.7, 1.0],
}
```

```python
## importing Stratified K Fold Cross Validation
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits = 10, shuffle=True, random_state=42)
```

```python
## importing random search cv
from sklearn.model_selection import RandomizedSearchCV
rcv = RandomizedSearchCV(
    estimator = rfc, 
    param_distributions = rfc_params, 
    scoring = 'accuracy', 
    cv = skf, 
    n_jobs = -1, 
    verbose = 2
)
```

```python
## fitting the model
rcv.fit(x_train, y_train)
```

```python
## checking cross validation score
scores = cross_val_score(estimator=rcv, X=x_train, y=y_train, scoring='accuracy', cv=skf, n_jobs=-1)
```

```python
## average cross validation score
scores.mean()
```

```python
## Randomized CV best scores
rcv.best_score_
```

```python
best = rcv.best_estimator_
```

```python
## fitting the dataset with best estimator for best results
best.fit(x_train, y_train)
```

```python
## making prediction with the testing dataset
y_pred_test = best.predict(x_test)
y_pred_train = best.predict(x_train)
```

```python
## evaluating the model with prediction data
from sklearn.metrics import f1_score, precision_score, accuracy_score
training, testing = [],[]

for met in [accuracy_score, precision_score, f1_score]:
    if met == accuracy_score:
        values_train = met(y_train, y_pred_train)    
        values_test = met(y_test, y_pred_test)
        training.append(values_train)
        testing.append(values_test)
    else:
        values_train = met(y_train, y_pred_train, average='micro')    
        values_test = met(y_test, y_pred_test, average='micro')
        training.append(values_train)
        testing.append(values_test)
```

```python
results = {
    'training' : training,
    'testing' : testing
}
```

```python
results_df = pd.DataFrame(data=results, index=['Accuracy', 'Precision', 'F1 Score'])
```

```python
results_df['difference'] = abs(results_df['training'] - results_df['testing'])
```

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
## initiallizing three algorithms and their parameters
model_params = [
    ('Gradient Boosting Classifier',
    GradientBoostingClassifier(),
     {
        'loss' : ['log_loss', 'exponential'],
        'learning_rate' : [0.1, 0.3, 0.5, 0.7, 0.9, 1.0],
        'n_estimators' : [100, 300, 500, 700, 900],
        'criterion' : ['friedman_mse', 'squared_error'],
        'min_samples_split' : [x for x in range(2, 10)],
        'min_samples_leaf': [x for x in range(1,10)],
        'max_depth': [x for x in range(0, 10)],
        'max_features' : ['sqrt', 'log2'],
     }
    ),
    ('AdaBoostClassifier',
    AdaBoostClassifier(),
     {
         'estimator': [None, DecisionTreeClassifier(max_depth=1), DecisionTreeClassifier(max_depth=3)],  # Base estimator
         'n_estimators': [50, 100, 200, 300, 500],  # Number of weak learners
         'learning_rate': [0.01, 0.1, 0.5, 1.0, 1.5],  # Weight applied to each classifier at each boosting iteration
         'algorithm': ['SAMME', 'SAMME.R'],  # Boosting algorithm type
         'random_state': [None, 42],  # Controls randomness
     }
    ),
    ('Extra Gradient Boosting',
    XGBClassifier(),
     {
        'n_estimators': [100, 300, 500, 700, 900],  # Number of boosting rounds
         'learning_rate': [0.01, 0.1, 0.3, 0.5, 0.7, 1.0],  # Boosting learning rate
         'max_depth': [3, 5, 7, 9, 11],  # Maximum depth of trees
         'gamma': [0, 0.1, 0.2, 0.3],
         'objective': ['binary:logistic', 'multi:softmax', 'multi:softprob'],  # Learning objective
         'booster': ['gbtree', 'gblinear', 'dart'],  # Booster type,
         'eval_metric': ['logloss', 'error', 'auc'],  
     }
    )
]
```

```python
unique_classes = np.unique(y_train)
class_mapping = {value: idx for idx, value in enumerate(unique_classes)}
y_train = np.array([class_mapping[val] for val in y_train])
```

```python
model_param = {}
for name, model, param in model_params:
    randomCV = RandomizedSearchCV(
        estimator = model, 
        param_distributions = param, 
        n_iter = 100, 
        verbose = 2, 
        cv = 5, 
        n_jobs = -1
    )
    randomCV.fit(x_train, y_train)
    model_param[name] = randomCV.best_params_

for model in model_param:
    print(f'-------------------------- Best Params for {model} --------------------------')
    print(model_param[model])
```

```python
models_with_parameters = {
    'Gradient Boosting Classifier' : GradientBoostingClassifier(n_estimators=100, min_samples_split=7, min_samples_leaf=3, max_features='log2', max_depth=4, loss='log_loss', learning_rate=0.5, criterion='friedman_mse'),
    'AdaBossting Classifier' : AdaBoostClassifier(random_state=None, n_estimators=500, learning_rate=1.0, estimator=DecisionTreeClassifier(max_depth=3), algorithm='SAMME'),
    'Extra Gradient Boosting Classifier' : XGBClassifier(objective='multi:softmax', n_estimators=100, max_depth=3, learning_rate=1.0, gamma=0.3, eval_metric='error', booster='dart')
}
```

```python
## let's make a function for evaluating model performance
def evaluate_models(true, predicted):
    accuracy = accuracy_score(true, predicted)
    precision = precision_score(true, predicted, average='micro')
    f1 = f1_score(true, predicted, average='micro')
    return accuracy, precision, f1
```

```python
# Initialize lists for storing results
new_model_list = []
new_r2_values = []
new_results = []

# Evaluate each model
for model_name, model in models_with_parameters.items():
    # Fit the model
    model.fit(x_train, y_train)

    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # Evaluate the model
    model_train_accuracy, model_train_precision, model_train_f1 = evaluate_models(y_train, y_train_pred)
    model_test_accuracy, model_test_precision, model_test_f1 = evaluate_models(y_test, y_test_pred)

    # Calculate R2 difference
    new_f1_difference = abs(model_train_f1 - model_test_f1)*100
    
    # Print results
    print(model_name)
    print('=' * 50)
    print('Model Performance for Training Set')
    print(f'-- Accuracy Score: {np.round(model_train_accuracy, 2)}')
    print(f'-- Precision Score: {np.round(model_train_precision, 2)}')
    print(f'-- F1 Score: {np.round(model_train_f1, 2)}')
    print('-' * 50)
    print('Model Performance for Testing Set')
    print(f'-- Accuracy Score:: {np.round(model_test_accuracy, 2)}')
    print(f'-- Precision Score: {np.round(model_test_precision, 2)}')
    print(f'-- F1 Score: {np.round(model_test_f1, 2)}')
    print('=' * 50)

    # Store results
    new_model_list.append(model_name)
    new_r2_values.append(model_test_f1)
    new_results.append((model_name, model_train_f1, model_test_f1, new_f1_difference))
```

### <span style='color:#6A9C89'>6. Additional Notes</span>

#### 1. Explain Bagging and Boosting methods. How is it different from each other.
##### Bagging and Boosting are two popular ensemble methods used to improve the performance of machine learning models.
##### Bagging 
    (Bootstrap Aggregating) involves training multiple models independently on different subsets of the training data, typically created by bootstrapping (sampling with replacement). The predictions from these models are then averaged (for regression) or voted on (for classification) to make the final prediction. The primary goal of bagging is to reduce the variance of the model, which can help prevent overfitting. An example of a bagging algorithm is Random Forest
##### Boosting 
    Boosting involves training models sequentially, where each new model tries to correct the errors made by the previous ones. The models are dependent on each other, and predictions are made by combining the weighted outputs of each model. Boosting aims to reduce bias and improve the model's accuracy. Popular boosting algorithms include AdaBoost and XGBoost. The key difference between bagging and boosting is that bagging works by reducing variance (through parallel training of independent models), while boosting reduces bias by focusing on correcting mistakes made by previous models in a sequential manner.

#### 2. Explain how to handle imbalance in the data.
    When dealing with imbalanced data, there are several strategies to improve model performance. One common approach is resampling, which involves either oversampling the minority class (e.g., using techniques like SMOTE to generate synthetic samples) or undersampling the majority class to create a balanced dataset. Another method is adjusting the class weights, where higher weights are assigned to the minority class during model training to make the model pay more attention to that class. Anomaly detection techniques, such as using One-Class SVM, can also be effective for treating the minority class as anomalies. Additionally, ensemble methods like Balanced Random Forest or EasyEnsemble create balanced subsets of the data to train multiple models. When evaluating models on imbalanced data, it's crucial to focus on metrics like Precision, Recall, F1-Score, and ROC-AUC, as accuracy can be misleading in such cases. These techniques and metrics help ensure that the model performs well even when the classes are imbalanced.
