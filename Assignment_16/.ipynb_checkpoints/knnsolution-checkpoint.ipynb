### <span style='color:#6A9C89'>Data Separation</span>

```python
df.head()
```

```python
## splitting the data into dependent and independent variables
x = df.drop(columns='type', axis=1)
y = df[['type']]
```

```python
## we dont need animal name, as it will make the dataset highly imbalance
x.drop(columns='animal name', inplace=True)
x.head(2)
```

```python
## let's split the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
## now let's standardize the dataset for better accuracy
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train) ### fit transforming the x_train
x_test_scaled = scaler.transform(x_test)       ### transforming the x_test
```

```python
## initializing K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

### <span style='color:#6A9C89'> Hyper-Parameter Tunning</span>

```python
## hyper-parameter tunning
KNeighborsClassifier()
params = {
    'n_neighbors':[x for x in range(1,5)],
    'weights':['uniform', 'distance'],
    'algorithm':['auto'],
    'p':[x*2 for x in range(1,5)] ,
    'metric':['minkowski', 'manhattan'],
}
```

```python
## Cross Validation for Data Validation
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
## Randomized Search CV for Hyper Parameter Tunning
from sklearn.model_selection import RandomizedSearchCV
rsc = RandomizedSearchCV(estimator=knn, param_distributions=params, scoring='accuracy', cv=skf)
## fitting RandomizedSearchCV
rsc.fit(x_train_scaled, y_train)
```

```python
## Best Parameter for the model
best_parameters = rsc.best_params_
print(f'The best parameters for model is\n {best_parameters}')
```

```python
## Best score
best_score = rsc.best_score_
print(f'The best score for model is {best_score*100}%')
```

```python
## assigning best estimator to a additional variables
best_model = rsc.best_estimator_
best_model
```

```python
## Let's check for the cross validation score
cross_validation_score = cross_val_score(estimator=best_model, 
                X=x,
                y=y, 
                scoring='accuracy', 
                cv=skf)
```

```python
print(f'Cross Validation Score for the Fitted Model is/n{cross_validation_score}')
```

```python
## getting the average of the cross validation of the fitted model
print(f'Average Cross Validation Score = {cross_validation_score.mean()}%')
```

### <span style='color:#6A9C89'> Evaluation</span>

```python
## Prediction for model
y_pred = best_model.predict(x_test_scaled)
```

```python
## Evaluation for the model
from sklearn.metrics import accuracy_score, precision_score, f1_score
accuracy = accuracy_score(y_pred, y_test)
precision = precision_score(y_pred, y_test, average='micro')
f1 = f1_score(y_pred, y_test, average='micro')
print(f'Accuracy Score = {accuracy}')
print(f'Precision Score = {precision}')
print(f'F1 Score = {f1}')
```

### <span style='color:#6A9C89'>Interview Question</span>

#### 1. What are the key hyperparameters in KNN?
The key hyperparameters in KNN are:

1. **Number of Neighbors (`k`)**: Determines how many nearest neighbors influence the prediction.
2. **Distance Metric (`metric`)**: Specifies how to calculate the distance (e.g., Euclidean, Manhattan, Minkowski).
3. **Weighting (`weights`)**: Defines whether all neighbors contribute equally or closer ones have higher influence (`uniform` or `distance`).
4. **Algorithm (`algorithm`)**: Chooses the search strategy (`auto`, `ball_tree`, `kd_tree`, `brute`).
5. **Leaf Size (`leaf_size`)**: Affects the efficiency of tree-based algorithms.

These control the model's complexity, efficiency, and accuracy.

#### 2. What distance metrics can be used in KNN?
1. **Euclidean Distance**: Straight-line distance between two points.  
   Formula: \( \sqrt{\sum (x_i - y_i)^2} \)
2. **Manhattan Distance**: Sum of absolute differences between points.  
   Formula: \( \sum |x_i - y_i| \)
3. **Minkowski Distance**: Generalized form of Euclidean and Manhattan distances.  
   Formula: \( (\sum |x_i - y_i|^p)^{1/p} \)
4. **Chebyshev Distance**: Maximum absolute difference along any coordinate.  
   Formula: \( \max(|x_i - y_i|) \)
5. **Hamming Distance**: Number of differing elements, used for categorical data.
6. **Cosine Similarity**: Measures angular difference rather than distance (useful for text data).    Formula: \( 1 - \frac{\textbf{A} \cdot \textbf{B}}{\|\textbf{A}\| \|\textbf{B}\|} \)
