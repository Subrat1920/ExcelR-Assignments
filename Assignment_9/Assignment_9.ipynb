```python
## let's grab all the columns name for a standard threshold
def corelated_cols(dataset, threshold):
    col_corr = set()  # Set to store column names with high correlation
    corr_matrix = dataset.corr()  # Calculate correlation matrix
    
    for i in range(len(corr_matrix.columns)):  # Loop through columns
        for j in range(i):  # Loop through rows before the diagonal
            if abs(corr_matrix.iloc[i, j]) >= threshold/100:  # Access with iloc
                col_name = corr_matrix.columns[i]  # Get the column name
                col_corr.add(col_name)  # Add the column name to the set
    return col_corr
```

```python
## let's check which columns are closely related that is more than 60%
corelated_cols(df, threshold=60)
```

### <span style='color:#6A9C89'>2. Dimensionality Reduction with PCA</span>

```python
## standardize the features to ensure they have a mean of 0 and a standard deviation.

### standardizing the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_df = scaler.transform(df)
```

```python
## converting the scaled data back into data frame as fit_transforms transforms the dataset into a numpy_array
scaled_df_temp = pd.DataFrame(scaled_df, columns = df.columns)
```

```python
## let's again check the histograms for ensuring standardization
plt.figure(figsize=(10,15))
plt.suptitle('Histograms for the datasets', fontsize=20)

for idx, col in enumerate(scaled_df_temp.columns, 1):
    plt.subplot(int(len(scaled_df_temp.columns)/2), 2, idx)
    sns.histplot(data=scaled_df_temp, x=col, kde=True)
    plt.title(f'Histogram for {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Frequency')
plt.tight_layout(rect=[0,0,1, 0.95])
plt.show()
```

### <span style='color:#DEAA79'>Conclusion</span>
##### We, can conclude that the dataset is shifted the mean to 0 in every columns

```python
## Determining the optimal number of principal component using techniques like scree plot or cumulative explained variance

## performing PCA
from sklearn.decomposition import PCA
pca = PCA()
pca_result = pca.fit(scaled_df)
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

### <span style='color:#DEAA79'>Scree Plot</span>

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, len(explained_variance_ratio)+1), explained_variance_ratio, marker='o', linestyle='--', color='r')
plt.title('Scree Plot for Original Dataset')
plt.xlabel('Principal Components')
plt.ylabel('Explained Varaince Ratio')
plt.xticks(range(1, len(explained_variance_ratio)+1))
plt.grid()
plt.show()
```

```python
# Cumulative Explained Variance
cumulative_variance = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='r')
plt.title('Cumulative Explained Variance for Original Dataset')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.95, color='b', linestyle='-', label='95% Variance Explained')  # Example: 95% threshold
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
## let's shrink the data to 95% of the actual data
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Optimal number of components: {optimal_components}")
```

```python
## so, the number of principal components decreases to 10 from 14 i.e., 95% shrinkage
## transforming the original dataset into the principal components

pca = PCA(n_components=10)
pca_df = pca.fit_transform(scaled_df)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,6))
plt.title('Sree Plot for PCA Components (10 Components)')
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.grid()
plt.show()
```

```python
## cummulative explained variance for principal components
cum_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10,6))
plt.title('Cummlative Diagram for 10 Principal Components')
plt.plot(range(len(cum_variance)), cum_variance, marker='o', linestyle='--')
plt.xlabel('Cummulative Variance')
plt.ylabel('Principal Components')
plt.grid()
plt.show()
```

### <span style='color:#6A9C89'>3. Clustering with Original Data</span>

```python
## let's transform the dataset
pca_result = pca.fit_transform(scaled_df)
```

```python
## let's see the behavious of features before clustering
plt.figure(figsize=(10,6))
plt.title('Features before Clustering')
plt.scatter(pca_result[:,0], pca_result[:,1])
plt.show()
```

```python
## applying k-means clustering
from sklearn.cluster import KMeans

## applying elbow method
### within cluster sum of square
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
```

```python
## plotting elbow method
plt.plot(range(1,11), wcss, marker='o', linestyle='--')
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

##### From the elbow curve we can somehow say that the number of clusters could be 3
##### But to be sure we can use another algorithm called kneelocator

```python
from kneed import KneeLocator
kl = KneeLocator(range(1,11), wcss, curve='convex', direction='decreasing')
print(f'The number of cluster could form is {kl.elbow}')
```

```python
## let's make the prediction to the original data
kmeans = KMeans(n_clusters=3, init='k-means++')
y_pred_original = kmeans.fit_predict(scaled_df)
```

```python
## clustering plot for original data
plt.figure(figsize=(10,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c = y_pred_original)
plt.title('Original Features after K-Means Clustering')
plt.show()
```

### <span style='color:#6A9C89'>4. Clustering with PCA Data</span>

```python
wcss_pca = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(pca_df)
    wcss_pca.append(kmeans.inertia_)
```

```python
## Plotting elbow curve for pca data
plt.figure(figsize=(10,6))
plt.title('Elbow Curve for Principal Components')
plt.plot(range(1,11), wcss, marker='*', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of Squares')
plt.grid()
plt.show()
```

```python
## clustering for the scaled data /  Principal Components
kmeans = KMeans(n_clusters=3, init='k-means++')
```

```python
y_pred_pca = kmeans.fit_predict(pca_df)
```

```python
## plotting the clustered plot
plt.figure(figsize=(10,6))
plt.title('Clustering Prinicipal Components with K-Means')
plt.scatter(pca_df[:,0], pca_df[:,1], c=y_pred_pca)
plt.show()
```

#### <span style='color:#DEAA79'>Silhoutte Scoring</span>

```python
from sklearn.metrics import silhouette_score
silhouette_score_list = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=10, init='k-means++')
    kmeans.fit(pca_df)
    score = silhouette_score(pca_df, kmeans.labels_)
    silhouette_score_list.append(score)
```

```python
## plotting silhouette elbow curve
plt.figure(figsize=(10,6))
plt.plot(range(2,11), silhouette_score_list, marker='o', linestyle='--')
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
```

#### 2. Similarities and Differences
##### 1. Similarities
    -> The general structure and separation of cluster are preserved inboth the original and PCA-Transformed datasets.
    -> The number of clusters and their spatial distribution are consistent acroos the two approaches.
##### 2. Differences
    -> PCA slighlty alters the shape of the clusters since it transforms the feature space to a lower-dimensional representation.
#### 3. Impact of Dimensionality Reduction on Clustering
##### 1. Preserved Variance: 
    PCA ensures that the most critical variance is retained, which helps maintain the clustering performance.
##### 2. Reduced Noise: 
    PCA can suppress noise and irrelevant features, potentially improving clustering accuracy.
##### 3. Simplified Visualization: 
    Clustering results in a reduced-dimensional space (e.g., 2D) are easier to visualize, as seen in the PCA-transformed plot.

#### Trade-Offs Between PCA and Clustering Directly on Original Data:
##### Advantages of PCA:
    Reduces the risk of overfitting, especially in high-dimensional datasets.
    Speeds up computations by working in a lower-dimensional space.
    Removes multicollinearity and redundant features.
##### Disadvantages of PCA:
    1. Information Loss: Although PCA preserves most variance, some minor details might be lost.
    2. Interpretability: Principal components are linear combinations of original features, making it harder to interpret results.
##### Advantages of Clustering on Original Data:
    1. Maintains full information from the dataset.
    2. Clustering results can be directly related to the original feature space.
##### Disadvantages of Clustering on Original Data:
    1. High-dimensional data can lead to sparsity issues and impact clustering performance.
    2. Noise and irrelevant features might distort cluster boundaries.

#### <span style='color:#DEAA79'>Practical Implications of Using PCA and Clustering</span>
##### Enhanced Efficiency:
    1. PCA reduces the computational cost by working with fewer features, which is particularly useful in high-dimensional datasets.
    2. Faster clustering processes in lower-dimensional spaces without significant loss of information.
    3. Improved Model Robustness:
    4. PCA removes multicollinearity and irrelevant noise, improving clustering accuracy and robustness.

#### <span style='color:#DEAA79'>Recommendations for Using PCA and Clustering:</span>
##### When to Use PCA:
    1. High-dimensional datasets with correlated features or potential multicollinearity.
    2. Situations where computational efficiency is critical.
    3. When visualization of the data in 2D or 3D is needed.
##### When to Use Clustering Directly:
    1. Low-dimensional datasets where all features are relevant and interpretable.
    2. When interpretability of feature space in clustering is essential (e.g., business applications where specific features drive decisions).
