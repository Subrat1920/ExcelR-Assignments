## <span style="color: #ffffff;background-color: #6A9C89;font-size: 1.2em;font-weight: bold;padding: 8px 12px;border-radius: 6px;box-shadow: 0px 2px 5px rgba(0, 0, 0, 0.2);texttransform: uppercase;letter-spacing: 1px;">Data Pre-Processing</span>

#### The important stage of association rule is that we need to convert the data into list format as apriori algorithm accepts the data in the form of list not in DataFrame.

```python
## let's check how many of the max total of items are there
first_transaction = str(df[0][0]) ## extracting only the first transaction
len(first_transaction.split(',')) ## assuming that in the first column there exists max number of columns.
```

```python
trans_list = []
for i in range(0, rows):
    trans = df[0][i].split(',')
    trans_list.append(trans)
```

```python
trans_list
```

#### So, we can see that we have pre-processed whold dataframe into association rule based data, removing null values.

## <span style="color: #ffffff;background-color: #6A9C89;font-size: 1.2em;font-weight: bold;padding: 8px 12px;border-radius: 6px;box-shadow: 0px 2px 5px rgba(0, 0, 0, 0.2);texttransform: uppercase;letter-spacing: 1px;">Association Rule Mining</span>

```python
## implementation of apriori algorithm
### importing apyori
from apyori import apriori 

## implementing apriori algorithm
association_rules = apriori(transactions=trans_list, 
                            min_support = 0.0023,
                            min_cofidence = 0.20,
                            min_lift = 3,
                            max_length = 2
                           )
```

```python
## changing the association rules into list, so that we can extract valuable information from this.
results = list(association_rules)
```

```python
## characteristics for the results
print(f'The length of the Result is {len(results)}')
print(f'So, we have {len(results)} rules based on the above dataset.')
```

#### So, from the above exploration we can find that we can extract the data from the list itself. So, let's create a table which will give us the whole summarization of the Association Rules that we have created, from which we can later retract insights.

#### <span style='color:#6A9C89'>Storing the statistical Result in a single data frame</span>

```python
## let's summarize the whole association rules for bettern understanding of the rules
### we can create a dataframe, or table to summarize the description from the rules
### let's create a null-dictionary which will containize the whole description of the rules we have created
rule_dict = []
for i in range(0, len(results)):
    ## for relational products and its support
    relational_Products = results[i][0]                                                  ### extract the name of the relational products
    support = results[i][1]                                                              ### support for the relational products

    ## for first base item 
    first_Base_Item = results[i][2][0][0]                                                ### LHS of the Rule First Rule/Item
    first_Add_Item = results[i][2][1][0]                                                 ### RHS of the Rule First Rule/Item
    first_Base_Item_Confidence = results[i][2][0][2]                                     ### Confidence
    first_Base_Item_Lift = results[i][2][0][3]                                           ### Lift

    ## for second base item 
    second_Base_Item = results[i][2][0][1]                                               ### LHS of the Rule Second Rule/Item
    second_Add_Item = results[i][2][1][1]                                                ### RHS of the Rule Second Rule/Item
    second_Base_Item_Confidence = results[i][2][1][2]                                    ### Confidence  
    second_Base_Item_Lift = results[i][2][1][3]                                          ### Lift

    ## appending them into the rule_dicationary
    rule_dict.append({
        'Relational Product' : relational_Products,
        'Support' : support,
        '1st Base Item' : first_Base_Item,
        '1st Add Item' : first_Add_Item,
        '1st Base Item Confidence' : first_Base_Item_Confidence,
        '1st Base Item Lift' : first_Base_Item_Lift,
        '2nd Base Item' : second_Base_Item,
        '2nd Add Item' : second_Add_Item,
        '2nd Base Item Confidence' : second_Base_Item_Confidence,
        '2nd Base Item Lift' : second_Base_Item_Lift
    })
```

```python
rule_df = pd.DataFrame(rule_dict)
rule_df
```

```python
## strip all the columns if there is some space in case
rule_df.columns = rule_df.columns.str.strip()
```

## <span style="color: #ffffff;background-color: #6A9C89;font-size: 1.2em;font-weight: bold;padding: 8px 12px;border-radius: 6px;box-shadow: 0px 2px 5px rgba(0, 0, 0, 0.2);texttransform: uppercase;letter-spacing: 1px;">Analyze and Interpretation</span>

```python
top_10_support_values = rule_df.sort_values(by='Support', ascending=False).head(10)
top_10_support_values
```

#### <span style='color:#6A9C89'>Graphical Represenatation for Sorted Support Values</span>

```python
## we should remove all the unexpected strings present in relational products string
top_10_support_values['Relational Product'] = top_10_support_values['Relational Product'].apply(lambda x: ' & '.join(sorted(x)))
```

```python
## support for each relational values
plt.figure(figsize=(10,6))
plt.title('Support Values for each Relational Products')
sns.barplot(data = top_10_support_values, y = 'Relational Product', x= 'Support', palette='Set2')
plt.ylabel('Relational Product')
plt.xlabel('Support Values')
plt.show()
```

```python
# Reshape data for better visualization
melted_df = pd.melt(top_10_support_values, 
                    id_vars=['1st Base Item', '2nd Base Item'], 
                    value_vars=['1st Base Item Confidence', '2nd Base Item Confidence'], 
                    var_name='Confidence Type', value_name='Confidence')

# Combine 'Base Item' columns into a single y-axis label
melted_df['Base Item'] = melted_df.apply(lambda row: row['1st Base Item'] if '1st' in row['Confidence Type'] 
                                         else row['2nd Base Item'], axis=1)

# Plot
plt.figure(figsize=(12, 8))
plt.title('Comparison of Confidence between Two Base Products')

# Barplot with hue for confidence type
sns.barplot(data=melted_df, x='Confidence', y='Base Item', hue='Confidence Type', palette='Set2')

plt.xlabel('Confidence')
plt.ylabel('Base Products')
plt.legend(title='Confidence Type')
plt.show()
```

#### <span style='color:#798645'> 1. What is lift and why is it important in Association Rules?</span>
    Lift is a metric used in association rule minning to measure the strength of an association between two tems, compared to what would be expected if the two items were independent. It evaluates whether the occurence of one item increases the likelihood of the other item occuring.



ALTERNATIVELY



    Where:
        1. Confidence (X → Y): Probability that Y occurs given X has occurred.
        2. Support (Y): Probability that Y occurs in the dataset.
    Interpreting Lift:
        1. Lift > 1:
            Indicates a positive association between X and Y.
            This means X and Y are more likely to occur together than by random chance.
        2. Lift = 1:
            Indicates that X and Y are independent.
            The occurrence of X does not affect the likelihood of Y.
        3. Lift < 1:
            Indicates a negative association between X and Y.
            This means X and Y are less likely to occur together than expected by chance.

#### <span style='color:#798645'>2.	What is support and Confidence. How do you calculate them?</span>
    Support and Confidence in Association Rules
##### 1. Support
        Support measures how frequently a particular itemset appears in the dataset. It helps identify how "popular" or "relevant" a rule is within the data.

    Interpretation:
        Higher support: Indicates the rule is common in the dataset.
        Lower support: Indicates the rule is rare and might not be significant for decision-making.

##### Confidence
    Confidence measures the likelihood of item 𝑌 being purchased given that item X is already purchased. It evaluates the strength of the implication X→Y.



    Interpretation:
        Higher confidence: Stronger belief that Y occurs when X occurs. Lower confidence: Weaker association between X and Y.

#### <span style='color:#798645'>3.	What are some limitations or challenges of Association rules mining?</span>
    1. Computational Complexity: Exponential growth in itemsets with large datasets makes mining expensive.
    2. Rare Items: Low support thresholds may exclude rare but valuable patterns.
    3. Too Many Rules: Generates a large number of rules, many of which may not be actionable.
    4. Correlation ≠ Causation: Rules show correlation, not causation, leading to potential misinterpretation.
    5. Handling Continuous Data: Requires discretization, which can cause loss of detail.
    6. Parameter Selection: Hard to set optimal support, confidence, and lift thresholds.
    7. Scalability: Struggles with large datasets without advanced algorithms or distributed computing.
    8. No Temporal Patterns: Ignores the sequence or timing of events.
    9. Subjective Rule Evaluation: Interestingness depends on domain expertise.
    10. Missing/Noisy Data: Reduces accuracy and reliability of results.
    11. Lack of Generalization: Rules may not apply across datasets or domains.
    12. Complex Rules: Hard-to-interpret rules with many antecedents limit practical use.
