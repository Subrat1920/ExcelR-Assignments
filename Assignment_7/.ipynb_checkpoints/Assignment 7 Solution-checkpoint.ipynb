```python
## let's remove the outliers in the required columns
def remove_outliers(data, columns, threshold=1.5):
    for i in columns:
        q1 = data[i].quantile(0.25)
        q3 = data[i].quantile(0.75)

        iqr = q3 - q1

        lower_whisker = q1 - threshold*(iqr)
        upper_whisker = q1 + threshold*(iqr)

        new_data = data[(data[i]>=lower_whisker) & (data[i]<=upper_whisker)]
        
        return new_data
```

```python
def corelated_cols(dataset, threshold):
    col_corr = set()  # Set to store column names with high correlation
    corr_matrix = dataset.corr()  # Calculate correlation matrix
    
    for i in range(len(corr_matrix.columns)):  # Loop through columns
        for j in range(i):  # Loop through rows before the diagonal
            if abs(corr_matrix.iloc[i, j]) >= threshold:  # Access with iloc
                col_name = corr_matrix.columns[i]  # Get the column name
                col_corr.add(col_name)  # Add the column name to the set
    return col_corr
```

### <span style='color:#6A9C89'>5. MODEL BUILDING</span>

```python
## splitting the training, testing dataset into independent and dependent variables
### TRAINING DATA
x_train = train_df.iloc[:,1:]   # dependent variables
y_train = train_df.Survived     # independent variables

### TESTING DATA
x_test = test_df.iloc[:,1:]      # dependent variables
```

```python
## standardizing the variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled_df = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled_df = pd.DataFrame(x_test_scaled, columns=x_test.columns)

# Plot before and after scaling
features = x_train.columns
fig, axes = plt.subplots(nrows=len(features), ncols=2, figsize=(12, len(features) * 5))

for i, feature in enumerate(features):
    # Before scaling
    axes[i, 0].hist(x_train[feature], bins=20, alpha=0.7, label='Train', color='blue')
    axes[i, 0].hist(x_test[feature], bins=20, alpha=0.7, label='Test', color='orange')
    axes[i, 0].set_title(f'{feature} - Before Scaling')
    axes[i, 0].legend()

    # After scaling
    axes[i, 1].hist(x_train_scaled_df[feature], bins=20, alpha=0.7, label='Train', color='green')
    axes[i, 1].hist(x_test_scaled_df[feature], bins=20, alpha=0.7, label='Test', color='red')
    axes[i, 1].set_title(f'{feature} - After Scaling')
    axes[i, 1].legend()

plt.tight_layout()
plt.show()
```

```python
## importing logistic regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
## importing stratified k folds for cross validation
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
params = {
    'penalty' : ['l1', 'l2', 'elasticnet'],
    'max_iter' : [100,200,400,500,600,900,1000],
    'solver' : ['lbfgs','sag', 'saga', 'newton-cg'],
    'n_jobs' : [-1]
}
```

```python
## importing grid search cv for hyper-parameter-tunning
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=log_reg, param_grid=params, scoring='accuracy', cv=skf)
```

```python
## fitting the model with x_train and y_train
grid.fit(x_train_scaled, y_train)
```

```python
print(f'Best Score for the model is {np.round(grid.best_score_, 2)}')
print(f'Best Params for the model is {grid.best_params_}')
print(f'Best Index for the model is {grid.best_index_}')
```

```python
## assinging best-estimator to new model
best_model = grid.best_estimator_
best_model
```

```python
## predicting the test-values using new model
y_pred = best_model.predict(x_test_scaled)
```

```python
output = pd.DataFrame(
    {
        'PassengerId':test_df['PassengerId'], 
        'Survived': y_pred
    }
)
```

```python
## cross-validation-score using stratified k fold
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(estimator=best_model, X=x_train, y=y_train,cv=skf)
print(f'Cross Validation Score of 10 items for best-model are \n {np.round(cvs,2)}')
print(f'The mean cross validation score is {np.round(cvs.mean(), 2)}')
```

### <span style='color:#6A9C89'>STREAMLIT APP CODE</span>

```python
## import neccesary packages and library
import streamlit as st
import pickle 
import numpy as np


## load the StandardScaler and Logistic Regression Pickle Models
model = pickle.load(open('model.pkl', 'rb')) # for logistic regression
scaler = pickle.load(open('scaler.pkl', 'rb')) # for standard scaler

## initializing streamlit app
def main():
    st.title('Titanic Survival Prediction')
    st.write('This app predicts wheter the passenger survived based on the data like Age, Pclass, Multiple Cabin, Sex, Siblings, Parental-Children, Fare Rate and Embarked')
    st.write('Please select the values for the features')


    ## form for input features
    
    pclass = st.number_input('Pclass')
    sex = st.number_input('Sex (female: 0 and male: 1)')
    age = st.number_input('Age')
    siblings = st.number_input('Siblings [0-5] ')
    parch = st.number_input('Parch [0-6]')
    fare = st.number_input('Fare')
    embarked = st.number_input('Emabrked [ C : 0, Q : 1, S : 2]')
    multiple_cabin = st.number_input('Multiple Cabin [0-4]')

    if st.button('Predict'):
        ## creating a list with all input features
        features = [pclass, sex, age, siblings, parch, fare, embarked, multiple_cabin]

        scaled_features = scaler.transform([features])
        
        result = model.predict(scaled_features)

        if result[0]==1:                  ## if passenger survived
            st.subheader('Passenger Survived')
        elif result[0] == 0:              ## if passanger died
            st.subheader('Passenger Died')
        else:                             ## input values are wrong
            st.subheader('Error in input values')

if __name__=='__main__':
    main()
```

### <span style='color:#6A9C89'>INTERVIEW QUESTIONS</span>

#### 1. What is the difference between precision and recall?
##### 1. Precision
    Definition: Precision measures the proportion of correctly predicted positive instances out of all instances that were predicted as positive.
    Precision focuses on how many of the predicted positive instances are actually correct.
    A high precision means fewer false positives.
##### 2. Recall
    Definition: Recall measures the proportion of correctly predicted positive instances out of all actual positive instances in the dataset.
    Recall focuses on how many of the actual positive instances are identified by the model.
    A high recall means fewer false negatives.

#### 2. What is cross-validation, and why is it important in binary classification?
Cross-validation is a vital technique in machine learning for evaluating model performance by partitioning the dataset into multiple training and testing sets. This iterative process ensures the model is tested on diverse subsets, reducing the likelihood of overfitting and bias. It provides a robust measure of the models ability to generalize to unseen data. In binary classification, cross-validation is crucial for assessing the models effectiveness, improving its reliability, and ensuring accurate predictions on both positive and negative classes in real-world scenarios.
